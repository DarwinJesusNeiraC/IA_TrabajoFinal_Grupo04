```python
import re # Biblioteca que permite trabajar con expresiones regulares
# Funcion mediante la cual podremos extraer el contenido de la columna entre []
def extract_column_content(column_name):
    # Utilizamos la siguiente expresi√≥n regular
    match = re.search(r'\[(.*?)\]', column_name)
    if match:
        return match.group(1)
    else:
        return column_name

data = pd.read_csv(careerUrl)
careers = pd.DataFrame(data)
# Transformamos los nombres de las columnas
careers.rename(columns=extract_column_content, inplace=True)
# Resultado
print(careers)
```

# **2. Implementaci√≥n del algoritmo KNN**

```python
# importamos las bibliotecas, requeridas para KNN, usando la
# biblioteca de aprendiza automatico sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Tabla de puntacion de los cursos de cada participante
X = table_values

# Carreras de los participantes del formulario
y = list_careers

# Normalizacion de los datos entre 0 y 1
scaler = MinMaxScaler()
X_normalized = scaler.fit_transform(X)

# Dividicion de datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.2, random_state=42)

# Crear un clasificador KNN con k=33
knn = KNeighborsClassifier(n_neighbors=33)

# Ajustamos el clasificador con los datos de entrenamiento
knn.fit(X_train, y_train)

# Realizamos predicciones usando el conjunto de prueba
y_pred = knn.predict(X_test)

# Evaluamos el rendimiento del modelo
accuracy = accuracy_score(y_test, y_pred)
print("Exactitud: ", accuracy)

# Reporte de clasificaci√≥n
print("Reporte de Clasificaci√≥n:")
print(classification_report(y_test, y_pred, zero_division=1))
```

# **4. Balanceo de Datos**

```python
# Veremos la cantidad de participantes de cada carrera
resultado_agrupado = careers.groupby('¬øQu√© carrera profesional elegiste?')

# Conteo de ocurrencias para cada carrera
conteo_carreras = resultado_agrupado.size()

# Carreras y la cantidad de veces que aparecen
print(conteo_carreras)
```

```python
# De la salida anterior podemos ver como "Ingenieria de Sistemas" tiene m√°s cantidad de datos
# que las dem√°s, por lo tanto se aplicara undersampling
condicion = (careers['¬øQu√© carrera profesional elegiste?'] == 'Ingenier√≠a De Sistemas')
indices_a_eliminar = careers[condicion].index[:11]
careers.drop(indices_a_eliminar, inplace=True)
```

```python
# Veremos el resultado de la aplicacion de undersampling
resultado_agrupado = careers.groupby('¬øQu√© carrera profesional elegiste?')

# Obtener el conteo de ocurrencias para cada carrera
conteo_carreras = resultado_agrupado.size()

# Imprimir las carreras y la cantidad de veces que aparecen
print(conteo_carreras)
```

```python
# Ahora aplicaremos oversampling al resto de carreras
conteo_por_carrera = careers['¬øQu√© carrera profesional elegiste?'].value_counts()

# Estableceremos cual sera la cantidad de datos que requerimos
maximo_registros_deseados = 10

# Creamos una lista vacia para almacenar las r√©plicas de filas
replicas = []

# Iteramos sobre las carreras y agrega replicas de filas seg√∫n sea necesario
for carrera, conteo_actual in conteo_por_carrera.items():
    if conteo_actual < maximo_registros_deseados:
        # Realizamos el calculo de r√©plicas necesitas para alcanzar 10 registros
        cantidad_replicas = maximo_registros_deseados - conteo_actual

        # Filtramos las filas que pertenecen a la carrera actual
        filas_carrera_actual = careers[careers['¬øQu√© carrera profesional elegiste?'] == carrera]

        # Generamos r√©plicas aleatorias de las filas de la carrera actual
        replicas_carrera_actual = filas_carrera_actual.sample(n=cantidad_replicas, replace=True)

        # Agregamos las r√©plicas a la lista de r√©plicas
        replicas.append(replicas_carrera_actual)

# Toda la data modificada sera guardada en balanced_careers
# Concatenamos las r√©plicas con careers
balanced_careers = pd.concat([careers] + replicas)

# Reiniciamos el √≠ndice de balanced_careers
balanced_careers.reset_index(drop=True, inplace=True)
```

```python
balanced_careers # Mostando contenido de balanced_careers
```

```python
# Veremos el resultado balanceado
resultado_agrupado = balanced_careers.groupby('¬øQu√© carrera profesional elegiste?')

# Obtener el conteo de ocurrencias para cada carrera
conteo_carreras = resultado_agrupado.size()

# Imprimir las carreras y la cantidad de veces que aparecen
print(conteo_carreras)
```

```python
# Actualizamos el valor de las variables table_values y list_careers
table_values = balanced_careers.iloc[:, :-1].values

X = table_values

list_careers = balanced_careers.iloc[:, -1].values

y = list_careers
```

```python
list_careers # Vemos el contenido actualizado de list_careers
```

# **5. Reentrenamiento del algoritmo KNN**

```python
# Normalizacion de los datos entre 0 y 1
scaler = MinMaxScaler()
X_normalized = scaler.fit_transform(X)

# Dividicion de datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.2, random_state=42)

# Crear un clasificador KNN con k=33
knn = KNeighborsClassifier(n_neighbors=33)

# Ajustamos el clasificador con los datos de entrenamiento
knn.fit(X_train, y_train)

# Realizamos predicciones usando el conjunto de prueba
y_pred = knn.predict(X_test)

# Evaluamos el rendimiento del modelo
accuracy = accuracy_score(y_test, y_pred)
print("Exactitud: ", accuracy)

# Reporte de clasificaci√≥n
print("Reporte de Clasificaci√≥n:")
print(classification_report(y_test, y_pred, zero_division=1))
```
